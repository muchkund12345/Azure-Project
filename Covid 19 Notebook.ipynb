```
vaccinationdata=vaccinationdata.withColumn("TOTAL_VACCINATIONS",col("TOTAL_VACCINATIONS").cast("int"))\
    .withColumn("TOTAL_VACCINATIONS_PER100",col("TOTAL_VACCINATIONS_PER100").cast("int"))\
    .withColumn("PERSONS_LAST_DOSE",col("PERSONS_LAST_DOSE").cast("int"))\
        .withColumn("NUMBER_VACCINES_TYPES_USED",col("NUMBER_VACCINES_TYPES_USED").cast("int"))\
            .withColumn("PERSONS_BOOSTER_ADD_DOSE",col("PERSONS_BOOSTER_ADD_DOSE").cast("int"))
```

```
vaccinationdata = vaccinationdata.withColumn("DATE_UPDATED", to_date("DATE_UPDATED", "yyyy-MM-dd"))\
    .withColumn("FIRST_VACCINE_DATE", to_date("FIRST_VACCINE_DATE", "yyyy-MM-dd"))
```

```
vaccinationdata = vaccinationdata.withColumn("PERSONS_VACCINATED_1PLUS_DOSE", col("PERSONS_VACCINATED_1PLUS_DOSE").cast(DecimalType(10, 10)))\
    .withColumn("PERSONS_VACCINATED_1PLUS_DOSE_PER100", col("PERSONS_VACCINATED_1PLUS_DOSE_PER100").cast(DecimalType(10, 10)))\
        .withColumn("PERSONS_LAST_DOSE_PER100", col("PERSONS_LAST_DOSE_PER100").cast(DecimalType(10, 10)))\
            .withColumn("PERSONS_BOOSTER_ADD_DOSE_PER100", col("PERSONS_BOOSTER_ADD_DOSE_PER100").cast(DecimalType(10, 10)))
```
